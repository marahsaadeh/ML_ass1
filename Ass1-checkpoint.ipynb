```python
#Remove it
def outliers(df,ft):
        Q1=df[ft].quantile(0.25)
        Q3=df[ft].quantile(0.75)
        IQR=Q3-Q1
        
        lower_bound=Q1-1.5*IQR
        upper_bound=Q3+1.5*IQR
        
        ls=df.index[(df[ft]<lower_bound)|(df[ft]>upper_bound)]
        return ls
```
